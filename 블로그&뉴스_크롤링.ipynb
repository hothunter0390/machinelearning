```python
#1page 가져오기

import requests
from bs4 import BeautifulSoup
```

```python

def get_news(URL):
  res = requests.get(URL)
  soup = BeautifulSoup(res.text, "html.parser") #파싱이라는 과정을 통해 문자로 되어있는 걸 컴퓨터가 이해할 수 있도록 하는 것


  title = soup.select_one("h2#title_area span").text #제목
  date = soup.select_one("span.media_end_head_info_datestamp_time")['data-date-time'] #기사작성일시, 클래스는 쩜
  media = soup.select_one("a.media_end_head_top_logo img")['title'] #매체명, a태그 밑에 있는 이미지 태그, 속성 타이틀
  content = soup.select_one("article#dic_area").text #기사원문
  content = soup.select_one("div#newsct_article").text.replace('\n','')

#함수로 묶기
  return(title, date, media, content, URL)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

keyword = "테슬라"
startdate = "2023.08.21"
enddate = "2023.09.21"


def get_news_list(keyword,startdate,enddate):
  URL = 'https://search.naver.com/search.naver?where=news&query={0}&sm=tab_opt&sort=2&photo=0&field=0&pd=3&ds={1}&de={2}&docid=&related=0&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so%3Ar%2Cp%3Afrom{3}to{4}&is_sug_officeid=0&office_category=0&service_area=0'.format(keyword, startdate, enddate, startdate.replace(".",""), enddate.replace(".",""))

  res = requests.get(URL)
  soup = BeautifulSoup(res.text, "html.parser")
  news_list = soup.select("ul.list_news li")


  li = []

  for item in news_list:
    if len(item.select("div.info_group a")) == 2:
      li.append(get_news(item.select("div.info_group a")[1]['href'])) #네이버 뉴스만을 더하라

  return pd.DataFrame(li, columns=['title', 'date', 'media', 'content', 'url'])
```
